```python
# 4. Ambil stopwords Bahasa Indonesia
indonesian_stop_words = set(stopwords.words('indonesian'))

# 5. Vektorisasi teks
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(stop_words=list(indonesian_stop_words))
X = vectorizer.fit_transform(df['note_content'])  # Perhatikan nama kolom
y = df['label']
```

```python
# 4. Augmentasi data menggunakan RandomOverSampler
ros = RandomOverSampler(random_state=42)
X_resampled, y_resampled = ros.fit_resample(X, y)
```

```python
# 5. Split data untuk training dan testing
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)
```

```python
# 6. Model 1: Random Forest
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
rf_pred = rf_model.predict(X_test)
print("\n--- Random Forest ---")
print(classification_report(y_test, rf_pred))
```

```python
# 7. Model 2: Logistic Regression
lr_model = LogisticRegression(max_iter=1000)
lr_model.fit(X_train, y_train)
lr_pred = lr_model.predict(X_test)
print("\n--- Logistic Regression ---")
print(classification_report(y_test, lr_pred))
```

```python
# 8. Model 3: Support Vector Machine
svm_model = SVC(kernel='linear')
svm_model.fit(X_train, y_train)
svm_pred = svm_model.predict(X_test)
print("\n--- SVM ---")
print(classification_report(y_test, svm_pred))
```
